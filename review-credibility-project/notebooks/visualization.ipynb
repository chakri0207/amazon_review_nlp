```python
# ================================
# 1. Import Libraries & Load Dataset
# ================================
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("../data/fake_reviews.csv")

# Map labels for readability
df['label_name'] = df['label'].map({'CG': 'Credible', 'OR': 'Fake'})

df.head()
```

```python
# ================================
# 2. Class Distribution Plot
# ================================
# Shows balance between Credible and Fake reviews

sns.countplot(x='label_name', data=df)
plt.title("Class Distribution: Credible vs Fake Reviews")
plt.xlabel("Review Type")
plt.ylabel("Count")
plt.show()
```

```python
# ================================
# 3. Review Length Comparison
# ================================
# Compare text length of credible vs fake reviews

df['review_length'] = df['text_'].apply(len)

sns.boxplot(x='label_name', y='review_length', data=df)
plt.title("Review Length Comparison")
plt.xlabel("Review Type")
plt.ylabel("Number of Characters")
plt.show()
```

```python
# ================================
# 4. Rating Distribution by Review Type (Optional)
# ================================
# Helps analyze rating behavior patterns

sns.countplot(x='rating', hue='label_name', data=df)
plt.title("Rating Distribution by Review Type")
plt.xlabel("Star Rating")
plt.ylabel("Count")
plt.show()
```

## ✅ Visualization Completed

- Class distribution confirms balanced dataset
- Review length shows textual behavior differences
- Rating distribution helps support credibility analysis
